*	MEDIANA O PERCENTIL 50: VALOR QUE SE ENUENTRA MITAD DE VALORES ENCIMA Y DEBAJO, NÚMERO PAR, SE TOMAN LOS DEL CENTRO Y PROMEDIAR. DE UNA LISTA ORDENADA DE MENOR A MAYOR.
*   ES ROBUSTA PORQUE NO ESTA INFLUENCIADA POR LOS VALORES ATÍPICOS (OUTLIERS) QUE PODRÍAN SESGAR LOS RESULTADOS.
*	EJEMPLO: INGRESOS FAMILIARES DE VARIOS BARRIOS. NO IMPORTA DE DONDE SEAN PORQUE NO SE CALCULA EL PROMEDIO, SE TOMA EL VALOR CENTRAL.

#### 1.3.2.1 VALORES ATÍPICOS

*	LOS VALORES ATÍPICOS NO NECESARIAMENTE SON ERRORES, A VECES SI COMO LA  MEDICIÓN DE UN SENSOR QUE FUNCIONE MAL, LA TENDENCIA CENTRAL SERÁ DEFICIENTE, MIENTRAS QUE LA MEDIANA NO.
*	LOS VALORES ATÍPICOS DEBEN IDENTIFICARSE, Y REALIZAR UNA INVESTIGACIÓN PROFUNDA. SE PUEDE UTILIZAR LA MEDIA TRUNCANDO EL 10% DE DATOS SUPERIORES E INFERIORES.

## 1.4-ESTIMACIÓN DE LA VARIABILIDAD O DISPERSIÓN


* MIDE EL GRADO DE DISPERSIÓN O AGRUPACIÓN DE LOS VALORES DE LOS DATOS. HAY QUE MEDIRLA, REDUCIRLA, DISTINGUIR LA VARIABILIDAD ALEATORIA DE LA REAL, IDENTIFICAR LAS DIVERSAS FUENTES DE LA VARIABILIDAD REAL Y TOMAR DECISIONES TENIÉNDOLA EN CUENTA.

### 1.4.1-DESVIACIÓN ESTÁNDAR Y ESTIMACIONES RELACIONADAS

* DESVIACIONES: LAS ESTIMACIONES DE LA VARIACIÓN MÁS UTILIZADAS SE BASAN EN LAS DIFERENCIAS ENTRE LAS ESTIMACIONES DE TENDENCIA CENTRAL Y LOS DATOS OBERVADOS. EJ: CONJUNTO DE DATOS {1, 4, 4}, LA MEDIA ES 3. LAS DESVIACIONES DE LA MEDIA SON LAS DIFERENCIAS: 1-3=-2 , 4-3=1 , 4-3= 1 ESTAS DESVIACIONES NOS INDICAN QUE TAN DISPERSOS ESTAN LOS DATOS DEL EN TORNO AL VALOR CENTRAL
* DESVIACIÓN MEDIA ABSOLUTA: UNA FORMA DE MEDIR LA VARIABILIDAD ES ESTIMAR EL VALOR TÍPICO PARA ESTAS DESVIACIONES. EXTRAER EL PROMEDIO DE LOS VALORES ABSOLUTOS DE LAS DESVIACIONES DE LA MEDIA. EN EL EJEMPLO ANTERIOR, LOS VALORES ABSOLUTOS DE LAS DESVIACIONES SON {2 1 1 } Y SU PROMEDIO ES 2+1+1=4/3=1,33. A ESTE VALOR SE LO CONOCE COMO DESVIACIÓN MEDIA ABSOLUTA.
* LAS ESTIMACIONES DE VARIABILIDAD MÁS CONOCIDAS SON: LA VARIANZA (PROMEDIO DEL CUADRADO DE LAS DESVIACIONES) Y LA DESVIACIÓN ESTANDAR (LA RAIZ CUADRADA DE LA VARIANZA). SE UTILIZA LA DESVIACIÓN ESTANDAR YA QUE ES MÁS CONVENIENTE EN LOS MODELOS ESTADÍSTICOS.
* LA DESVIACIÓN ESTANDAR Y LA VARIANZA NO SON ROBUSTAS.  LA DESVIACIÓN ABSOLUTA MEDIANA (MAD: MEDIANA DE LOS VALORES ABSOLUTOS DE LAS DESVIACIONES DE LA MEDIANA) SI ES ROBUSTA, AL IGUAL QUE LA DESVIACIÓN ESTANDAR TRUNCADA ANALOGA A LA MEDIA TRUNCADA. 

### 1.4.2-ESTIMACIÓN BASADA EN PERCENTILES

* CON LOS DATOS ORDENADOS, BUSCAR POR EJEMPLO EL PERCENTIL 80, COMENZAR POR EL VALOR MÁS PEQUEÑO HASTA LLEGAR AL 80% DE LOS DATOS. ES LO MISMO QUE EL CUANTIL, POR LO QUE EL CUANTIL 0.8 ES IGUAL AL PERCENTIL 80.
* LA MEDIDA HABITUAL ES EL RANGO INTERCUARTÍLICO (IQR): LA DIFERENCIA ENTRE EL PERCENTIL 25 Y 75. EJEMPLO: {3,1,5,3,6,7,2,9}, LOS ORDENAMOS {1,2,3,3,5,6,7,9}. EL PERCENTIL 25 ESTA EN 2,5 Y EL PERCENTIL 75 ESTA EN 6,5 POR LO QUE EL RANTO INTERCUARTÍLICO ES 6,5-2,5=4.

### 1.5.2 TABLAS DE FRECUENCIAS E HISTOGRAMAS

* La tabla de frecuencias de una variable divide el rango (La diferencia entre el mayor y el menor valor de un conjunto de datos) de la variable en segmentos igualmente espaciados y nos dice cuántos valores caen dentro de cada segmento

```python
## La función cut de pandas crea intervalos de igual valor. El método value_counts, devuelve una tabla de frecuencias.
binnedPopulation = pd.cut(state['Population'], 10)
print(binnedPopulation.value_counts())
```

```python
## Para crear la tabla debemos realizar:
# Table 1.5 #Tabla de frecuencias de la población por estados
binnedPopulation.name = 'binnedPopulation'
df = pd.concat([state, binnedPopulation], axis=1)
df = df.sort_values(by='Population')

groups = []
for group, subset in df.groupby(by='binnedPopulation'):
    groups.append({
        'Rango': group,
        'Recuento': len(subset),
        'Estados': ','.join(subset.Abbreviation)
    })
print(pd.DataFrame(groups))
```

El estado menos poblado es WY Wyoming con 526.935 habitantes y el más poblado es CA California con 37.253.956 habitantes. Esto nos da un rango de 37.253.956 - 526.935 = 36.727.021, que debemos dividir en contenedores de igual tamaño, por ejemplo 10. Cada uno tendrá 36.727.021/10 = 3.672.702,1. Hay que tener en cuenta los contenedores que estan vacios que nos brindan información.

* El histograma es una forma de visualizar la tabla de frecuencias, los contenedores en el eje x y los valores de los datos en el eje y

```python
# Histograma de la población por estados.
ax = (state['Population'] / 1_000_000).plot.hist(figsize=(4, 4))
ax.set_xlabel('Population (millions)')

plt.tight_layout()
plt.show()
```

NOTA: En teoría estadística, la localización (tendencia central) y variabilidad (dispersión) se conocen como primer y segudo momentos de la distribución. Asimetría (sí los datos estan sesgados hacia valores mayores o menores) y Curtosis (indica propensión de los datos a tener valores extremos) son el tercer y cuarto momento.

* Relacionado con el histograma existe el diagrama de densidad, muestra la distribución de los valores de los datos mediante una línea continua.

## 1.7 - CORRELACIÓN

En muchos proyectos de ciencia de datos o de investigación, implica examinar la correlación entre variables predictoras y entre variable predictora y variable objetivo. Las variables X e Y estan correlacionadas positivamente si los valores altos de X acompañan a los valores altos de Y o si sudece lo mismo con los valores bajos. Las variables X e Y estan correlacionadas negativamente si los valores altos de X acompañan a los valores bajos de Y o viceversa. 
Coeficiente de correlación: Métrica que mide el grado en el que las variables numéricas están asociadas entre sí (varía de -1 a 1). Utilizamos el coeficiente de correlación de Pearson, cuanto más cerca de -1 más correlación negativa, 0 no hay correlación y más cerca de 1 correlación positiva. ESTAS CORRELACIONES SON LINEALES. Si las variables tienen una asociación no lineal, esta métrica no es útil.

### 1.8.2 - DOS VARIABLES CATEGÓRICAS

Una forma conveniente es mediante una tabla de contigencia (tabla cruzada o como tabla de dos vías), una tabla de recuento por categorías. 
La tabla 1.8 muestra la tabla de contingencia con la calificación de un préstamo personal y el resultado de ese préstamo. La calificación va desde A a G. El resultado es: totalmente pagado, al corriente de pago, atrasado o cancelado.

```python
#Generar el dataframe lc_loans
LC_LOANS_CSV = "C:/Users/Admin/OneDrive/Big Data/Estadística/Práctica-Estadística/Data/lc_loans.csv"
lc_loans = pd.read_csv(LC_LOANS_CSV)
```

```python
# Tabla 1.8 - Tabla de contingencia de las calificaciones y estados de los préstamos de Lending Club
crosstab = lc_loans.pivot_table(index='grade', columns='status', 
                                aggfunc=lambda x: len(x), margins=True)
# Método pivot_table de Pandas para crear una tabla de resumen cruzado (crosstab) a partir de dos variables, grade y status
# index indica que se uliza grade como índice, columns columna de la tabla
# len (x) cuenta la cantidad
# margins se agregan columnas con los totales
print(crosstab)
```

```python
#Tabla 1.8 con porcentajes
df = crosstab.copy().loc['A':'G',:]
# .copy crea una copia seleccionando las filas de la A a la G y todas las columnas
df.loc[:,'Charged Off':'Late'] = df.loc[:,'Charged Off':'Late'].div(df['All'], axis=0)
# Toma las columnas hasta All y luego divide por la columna all axis = 0 indica por fila
df['All'] = df['All'] / sum(df['All'])
# Esto normaliza la columna all dividiendo cada valor por la suma total de la columna all, esto convierte a la columna all en porcentajes relativos en relación con el total.
perc_crosstab = df
print(perc_crosstab)
```

Observaciones: Los préstamos de calificación alta el porcentaje de retraso es bajo mientras los de calificación baja es más alto.

### 1.8.4 - VISUALIZACIÓN DE VARIAS VARIABLES

Los tipos de gráfico que se utilizan para comparar dos variables, diagramas de dispersión, agrupación hexagonal y diagrama de cajas, se extiende a más variables mediante la noción de acondicionamiento.
Recordemos la figura 1.8 que muestra la relación entre pies cuadrados terminados y sus valores de tasación fiscal, observamos que un grupo de viviendas que tienen un valor fiscal más alto.
La figura 1.12 explica el efecto de la localización al representar los datos para un conjunto de códigos postales.  

```python
# Figura 1.12
# Separar en un dataframe los cp de interes
zip_codes = [98188, 98105, 98108, 98126]
kc_tax_zip = kc_tax0.loc[kc_tax0.ZipCode.isin(zip_codes),:]
kc_tax_zip.head()
```

```python
#Esta función se utiliza para crear un gráfico de dispersión hexagonal (hexbin) y personalizar su apariencia.
def hexbin(x, y, color, **kwargs):
#**kwargs permite pasar argumentos adicionales que se utilizarán en la función.
    cmap = sns.light_palette(color, as_cmap=True)
#sns.light_palette() utiliza este color como base para generar una paleta de colores.
# as_cmap=True actue como mapa de color
    plt.hexbin(x, y, gridsize=25, cmap=cmap, **kwargs)
#plt.hexbin crear el gráfico hexagonal

g = sns.FacetGrid(kc_tax_zip, col='ZipCode', col_wrap=2)
#sns.FacetGrid se utiliza para crear un conjunto de gráficos enrejados (grid) para cada uno de los códigos postales en kc_tax_zip.
g.map(hexbin, 'SqFtTotLiving', 'TaxAssessedValue', 
      extent=[0, 3500, 0, 700000])
#g.map() se utiliza para asignar la función hexbin a cada gráfico del conjunto.
#extent establece el rango de los ejes X e Y para cada gráfico en el conjunto.
g.set_axis_labels('Superficie terminada en pies cuadrados', 'Valor de tasación fiscal')
g.set_titles('Código Postal {col_name:.0f}')

plt.tight_layout()
plt.show()
```

Observaciones: el valor fiscal es más alto en el cp 98126 y 98105, esta disparidad da lugar a los aclomerados de la figura 1.8
